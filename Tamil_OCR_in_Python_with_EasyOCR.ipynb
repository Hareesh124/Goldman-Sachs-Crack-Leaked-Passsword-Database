## Display Tamil Image

```
# show an image
import PIL
from PIL import ImageDraw
im = PIL.Image.open("tlpy1.jpg")
im
```

### Download Tamil and English Models 

```
# Create a reader to do OCR.
# If you change to GPU instance, it will be faster. But CPU is enough.
# (by MENU > Runtime > Change runtime type > GPU, then redo from beginning )
import easyocr
reader = easyocr.Reader(['te','en'])
```

## Perform OCR and extract Bounding Boxes

```
# Doing OCR. Get bounding boxes.
bounds = reader.readtext('tlpy1.jpg')
l =[]
# bounds
# len(bounds)-2
# bounds[0][1]
for i in range(len(bounds)):
  b = bounds[i][1]
  l.append(b)
# print(l)
s = ''
for i in range(len(l)):
  s = s+' '+l[i]
print(s)
```

```
!pip install googletrans==3.1.0a0
```

```
from googletrans import Translator
translator = Translator()
translations = translator.translate(l, dest='en')
res = ''
for translation in translations:
  res = res + ' ' + translation.text
print(res)
```

```
from gtts import gTTS #Import Google Text to Speech
from IPython.display import Audio #Import Audio method from IPython's Display Class
tts = gTTS(s) #Provide the string to convert to speech
tts.save('1.wav') #save the string converted to speech as a .wav file
sound_file = '1.wav'
Audio(sound_file, autoplay=True) 
```

### Draw the extracted Bounding boxes

```
# Draw bounding boxes
def draw_boxes(image, bounds, color='blue', width=2):
    draw = ImageDraw.Draw(image)
    for bound in bounds:
        p0, p1, p2, p3 = bound[0]
        draw.line([*p0, *p1, *p2, *p3, *p0], fill=color, width=width)
    return image

draw_boxes(im, bounds)
```
